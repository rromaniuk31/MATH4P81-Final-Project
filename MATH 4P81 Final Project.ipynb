## Create Simple Random Sample Without Replacement

```python
df['y-ybar'] = ''
df['y-ybar_sq'] = ''
for i in range(0,268):
    df['y-ybar'][i] = df['PF**'][i]-mean
    df['y-ybar_sq'][i] = df['y-ybar'][i]**2

df
```

```python
s_sq_numerator = df['y-ybar_sq'].sum()
S_sq = s_sq_numerator/(count-1)
S_sq
```

```python
#Athletic Director wishes to have a standard error of no greater than 4
n_SRS = S_sq/(4+(S_sq/count))
n_SRS = math.ceil(n_SRS)
n_SRS
```

```python
SRSdf['y-ybar'] = ''
SRSdf['y-ybar_sq'] = ''
for i in range(1,33):
    SRSdf['y-ybar'][i] = SRSdf['Points'][i] - SRSmean
    SRSdf['y-ybar_sq'][i] = SRSdf['y-ybar'][i]**2
    
s_sq = (SRSdf['y-ybar_sq'].sum())/(SRScount-1)

var_ybar = ((1-SRScount/count)*s_sq)/SRScount
var_ybar
```

## Normality Test

```python
## Shapiro-Wilk

data = SRSdf['Points']
from scipy.stats import shapiro
stat, p = shapiro(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print("Data follows Normal Distribution")
else:
    print("Data does not follow Normal Distribution")
```

## Create Stratified Random Sample w/ Proportional Allocation

```python
random.seed(2021)
```

```python
df['SeedStratum'] = ''

for i in range(0,268):
    if df['Seed'][i] <6:
        df['SeedStratum'][i] = "1-5"
    elif df['Seed'][i] <11:
        df['SeedStratum'][i] = "6-10"
    elif df['Seed'][i] >10:
        df['SeedStratum'][i] = "11-16"
        
df
```

```python
count_top5 = 0
count_6to10 = 0
count_bottom6 = 0

for i in range(0,268):
    if df['SeedStratum'][i] == "1-5":
        count_top5 += 1
    elif df['SeedStratum'][i] == "6-10":
        count_6to10 += 1
    elif df['SeedStratum'][i] == "11-16":
        count_bottom6 += 1
        
print("Top 5 = {}\n".format(count_top5), "Middle 5 = {}\n".format(count_6to10), "Bottom 6 = {}\n".format(count_bottom6))
```

```python
weight_top5 = count_top5/count
weight_6to10 = count_6to10/count
weight_bottom6 = count_bottom6/count

print("Top 5 Weight = {}\n".format(weight_top5), "Middle 5 Weight = {}\n".format(weight_6to10), "Bottom 6 Weight = {}\n".format(weight_bottom6))
```

```python
#Use n = 32 from simplified taking 11.94% of N
n_top5 = weight_top5*32
n_6to10 = weight_6to10*32
n_bottom6 = weight_bottom6*32

# we will round to the nearest integer so 15, 8, 9 respectively
print("n Top 5 = {}\n".format(n_top5), "n Middle 5 = {}\n".format(n_6to10), "n Bottom 6 = {}\n".format(n_bottom6))
```

```python
Strat_top5 = df[df.SeedStratum == '1-5']
Strat_top5.reset_index(drop=True, inplace=True)

Strat_6to10 = df[df.SeedStratum == '6-10']
Strat_6to10.reset_index(drop=True, inplace=True)

Strat_bottom6 = df[df.SeedStratum == '11-16']
Strat_bottom6.reset_index(drop=True, inplace=True)
```

## Create Stratified Random Sample w/ Neyman Allocation

```python
random.seed(2019)
```

```python
Strat_top5['y-ybar'] = ''
for i in range(0,122):
    Strat_top5['y-ybar'][i] = Strat_top5['PF**'][i]-ybar_top5
    Strat_top5['y-ybar_sq'][i] = Strat_top5['y-ybar'][i]**2
S1_sq = (Strat_top5['y-ybar_sq'].sum())/(Strat_top5['PF**'].count()-1)


Strat_6to10['y-ybar'] = ''
for i in range(0,67):
    Strat_6to10['y-ybar'][i] = Strat_6to10['PF**'][i]-ybar_6to10
    Strat_6to10['y-ybar_sq'][i] = Strat_6to10['y-ybar'][i]**2
S2_sq = (Strat_6to10['y-ybar_sq'].sum())/(Strat_6to10['PF**'].count()-1)


Strat_bottom6['y-ybar'] = ''
for i in range(0,79):
    Strat_bottom6['y-ybar'][i] = Strat_bottom6['PF**'][i]-ybar_bottom6
    Strat_bottom6['y-ybar_sq'][i] = Strat_bottom6['y-ybar'][i]**2
S3_sq = (Strat_bottom6['y-ybar_sq'].sum())/(Strat_bottom6['PF**'].count()-1)

S1 = math.sqrt(S1_sq)
S2 = math.sqrt(S2_sq)
S3 = math.sqrt(S3_sq)
print('S1_sq = {} , S1 = {}\n'.format(S1_sq,S1), 'S2_sq = {} , S2 = {}\n'.format(S2_sq,S2), 'S3_sq = {} , S3 = {}\n'.format(S3_sq,S3))
```

```python
n1_ney = (n_SRS * count_top5*S1)/(count_top5 * S1 + count_6to10 * S2 + count_bottom6 * S3)
n2_ney = (n_SRS * count_6to10*S2)/(count_top5 * S1 + count_6to10 * S2 + count_bottom6 * S3)
n3_ney = (n_SRS * count_bottom6*S3)/(count_top5 * S1 + count_6to10 * S2 + count_bottom6 * S3)
print('n1 neyman = {}\n'.format(n1_ney), 'n2 neyman = {}\n'.format(n2_ney), 'n3 neyman = {}'.format(n3_ney))
```
